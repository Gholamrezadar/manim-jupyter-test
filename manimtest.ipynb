```python
from manim import *
import manim

config.media_width = "75%"
config.verbosity = "WARNING"
```

```python
%%manim -qh Scene_RGB

class Scene_RGB(Scene):
    def construct(self):
        
        # RGB Text
        text = Text("RGB", fill_opacity=0.0, stroke_width=2, font_size=300)
        self.wait(0.2)
        self.play(FadeIn(text), run_time=3)
        self.wait(0.1)
        self.play(
            text.animate.shift(UP*3.2).scale(0.25).set_fill(color=WHITE, opacity=1),
            run_time=0.8
        )

        self.wait(0.5)

        # origin_dot = Dot(ORIGIN)
        # debug_circle = Circle(1, color=GRAY)
        # self.add(origin_dot, debug_circle)

        # numberplane = NumberPlane()
        # self.add(numberplane)

        # Circles
        RADIUS = 1.5
        circle_r = Circle(RADIUS, color=RED).shift(UP*RADIUS).rotate_about_origin(0*2*PI/3).shift(DOWN)
        circle_g = Circle(RADIUS, color=GREEN).shift(UP*RADIUS).rotate_about_origin(1*2*PI/3).shift(DOWN)
        circle_b = Circle(RADIUS, color=BLUE).shift(UP*RADIUS).rotate_about_origin(2*2*PI/3).shift(DOWN)

        self.play(Create(circle_r))
        # self.wait(0.1)
        self.play(Create(circle_g))
        # self.wait(0.1)
        self.play(Create(circle_b))

        self.wait(0.3)

        self.play(
            circle_r.animate.set_fill(opacity=0.33), 
            circle_g.animate.set_fill(opacity=0.33), 
            circle_b.animate.set_fill(opacity=0.33),
            run_time=0.5
            )

        

        self.wait(0.5)


```

```python
%%manim -ql --format gif Scene_1

class Scene_1(Scene):
    def construct(self):
        # Circle 1
        circle_1 = Circle(color=RED)
        circle_1.scale(3)
        circle_1.set_fill(RED, opacity=1)
        
        # Circle 2
        circle_2 = Circle(color=GREEN)
        circle_2.scale(2)
        circle_2.set_fill(GREEN, opacity=1)
        
        # Circle 3
        circle_3 = Circle(color=BLUE)
        circle_3.set_fill(BLUE, opacity=1)

        self.play(Create(circle_1, 2))
        self.play(Create(circle_2, 2))
        self.play(Create(circle_3, 2))
```
